```python
def outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Location'):
        m = np.mean(subdf.Price_per_sqft)
        st = np.std(subdf.Price_per_sqft)
        reduced_df = subdf[(subdf.Price_per_sqft>(m-st)) & (subdf.Price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

```python
df4 =outliers(df3)
df4.shape
```

```python
dummy= pd.get_dummies(df5.Location)
dummy.head()
```

```python
df6= pd.concat([df5,dummy.drop("other",axis="columns")],axis= "columns")
df6.head()
```

```python
df7= df6.drop("Location",axis= "columns")
```

```python
df7.head()
```

```python
x= df7.drop(["Price"],axis= "columns")
label= df7.Price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,label,test_size= 0.2,random_state= 10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
```

```python
cv= ShuffleSplit(n_splits= 5,test_size= 0.2,random_state= 0)
cross_val_score(LinearRegression(),x,label,cv= cv)
```

```python
def pred_price(area,rooms,gym,indoor,gas,location):
    temp= np.zeros(len(x.columns))
    loc_index= np.where(x.columns==location)[0][0]
    temp[0]= area
    temp[1]=rooms
    temp[2]=gym
    temp[3]=indoor
    temp[4]=gas
    temp[loc_index]=1
    return lr.predict([temp])[0]
    
    
```
